```python
%pip install pandas seaborn matplotlib datasets adjustText


import pandas as pd
import seaborn as sns
from datasets import load_dataset  
import matplotlib.pyplot as plt
 
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_DA= df[df['job_title_short'] == 'Data Analyst'].dropna(subset=['salary_year_avg'])
df_DA.loc[:, 'job_skills'] = df_DA['job_skills'].apply(lambda x: x.split(',') if isinstance(x, str) else x)


# Now explode the column
df_DA = df_DA.explode('job_skills')
df_DA["job_skills"] = df_DA["job_skills"].str.replace(r"[\[\]]", "", regex=True)
df_DA[['job_skills', 'salary_year_avg']]
```

```python
df_DA_skills =df_DA.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='count', ascending=False)
df_DA_skills = df_DA_skills.rename(columns={'count': 'skill_count', 'median':'median_salary'})
length = len(df_DA)
df_DA_skills['df_skills_perc']= 100 * df_DA_skills['skill_count']/length 
skills_perc = 5
df_DA_skills.head(30)
```

```python
import matplotlib.pyplot as plt
from adjustText import adjust_text
from matplotlib.ticker import FuncFormatter

# Assuming df_DA_skills is your DataFrame containing 'skill_count' and 'median_salary'
df_DA_skills.plot(kind='scatter', x='skill_count', y='median_salary')

texts = []
for i, text in enumerate(df_DA_skills.index):
    texts.append(plt.text(df_DA_skills['skill_count'].iloc[i], 
                          df_DA_skills['median_salary'].iloc[i], 
                          str(df_DA_skills['skill_count'].iloc[i]), ha='center'))

# Adjust text to avoid overlapping
adjust_text(texts, arrowprops=dict(arrowstyle='->', color='gray'))

plt.xlabel('Yearly salary (USD)')  # x-axis label
plt.ylabel('Skill count')  # y-axis label (assuming you're plotting salary against skills count)
plt.title('Most Optimal Skills for Data Analyst')

# Formatting the x-axis to display salaries in thousands of USD
ticks_x = FuncFormatter(lambda x, pos: f'${int(x/1000)}k')
plt.gca().xaxis.set_major_formatter(ticks_x)

plt.show()





```
