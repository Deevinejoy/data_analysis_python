```python
%pip install pandas seaborn matplotlib datasets


import pandas as pd
import seaborn as sns
from datasets import load_dataset  
import matplotlib.pyplot as plt
 
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_US = df[df['job_country'] == 'United States'].dropna(subset=['salary_year_avg'])
df_US
```

```python
job_titles= df_US['job_title_short'].value_counts().index[:10].tolist()
df_US_10 = df_US[df_US['job_title_short'].isin(job_titles)]
df_US_10 
```

```python

sns.boxplot(data=df_US_10, x='salary_year_avg', y='job_title_short')
sns.set_theme(style='ticks')
sns.despine()
plt.xlabel('Yearly salary (USD)')  # x-axis label
plt.ylabel('Job title')  # y-axis label
plt.title('Salary distribution of united states')
plt.xlim(0, 600000)
ticks_x = plt.FuncFormatter(lambda x, post: f'${int(x/1000)}k')
plt.gca().xaxis.set_major_formatter(ticks_x)
plt.show()


```

```python
df_DA = df[df['job_title_short'] == 'Data Analyst']
df_DA.loc[:, 'job_skills'] = df_DA['job_skills'].apply(lambda x: x.split(',') if isinstance(x, str) else x)


# Now explode the column
df_DA = df_DA.explode('job_skills')
df_DA["job_skills"] = df_DA["job_skills"].str.replace(r"[\[\]]", "", regex=True)
df_DA[['job_skills', 'salary_year_avg']].head(10)
df_DA = df_DA.dropna(subset=['salary_year_avg'])
```

```python
df_DA_skills_count =df_DA.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='count', ascending=False).head(6)
df_DA_skills_count
```

```python
df_DA_skills_med = df_DA.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='median', ascending=False).head(6)
df_DA_skills_med
```

```python
fig, ax = plt.subplots(2, 1)
sns.barplot(data=df_DA_skills_count, x='count', y=df_DA_skills_count.index, hue='count', ax=ax[0], palette='light:b')
sns.set_theme(style='ticks')
sns.despine()
ax[0].set_xlabel('Job count per skill') # x-axis label
ax[0].set_ylabel('')  # y-axis label
ax[0].legend().remove()

ax[0].set_title('top 10 most in-demand skills for data analysis')



sns.barplot(data=df_DA_skills_med, x='median', y=df_DA_skills_med.index,  hue='median', ax=ax[1], palette='light:b')
sns.set_theme(style='ticks')
sns.despine()
ax[1].set_xlabel('Yearly salary')  # x-axis label
ax[1].set_ylabel('')  # y-axis label
ax[1].legend().remove()
ax[1].set_title(' top 10 highest paid skills for data analysis')
ticks_x = plt.FuncFormatter(lambda x, post: f'${int(x/1000)}k')
ax[1].xaxis.set_major_formatter(ticks_x)
plt.tight_layout()
```
